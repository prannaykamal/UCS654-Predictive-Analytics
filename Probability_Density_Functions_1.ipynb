```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/data.csv",encoding="latin1")
df
```

```python
r=102317125
ar=0.05*(r%7)
br=0.3*(r%5+1)
print(ar)
print(br)
```

```python
x=df['no2']
x=x.dropna().values
z=x+ar*np.sin(br*x)
z
```

```python
mean = z.mean()
print(mean)
```

```python
lambda1 = len(z)/(2*((z-mean)**2).sum())
print(lambda1)
```

```python
c = np.sqrt(lambda1/np.pi)
print(c)
```
